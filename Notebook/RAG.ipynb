```python
genai.configure(api_key="AIzaSyBKhTXbpfX9fBBGW6gEDKn5dc_ez-s7hWw")
model = genai.GenerativeModel('gemini-1.5-flash')

# Step 2: Create a wrapper class
class GeminiEmbeddingModel:
    def __init__(self, model):
        self.model = model

    def embed_documents(self, documents):
        embeddings = []
        for doc in documents:
            embedding = self.model.get_embedding(doc.text)
            embeddings.append(embedding)
        return embeddings

    def embed_query(self, query):
        return self.model.get_embedding(query)

# Step 3: Use the model with VectorStoreIndex
gemini_embed_model = GeminiEmbeddingModel(model)
index = VectorStoreIndex.from_documents(documents, embed_model=gemini_embed_model, show_progress=True)
```
